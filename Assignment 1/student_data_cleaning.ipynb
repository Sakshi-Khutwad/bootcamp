```python
import pandas as pd # type: ignore

df = pd.read_csv(r"./Students.csv")

df.isnull().sum()
```

```python

# Handling the missing data by replacing missing MathScore and EnglishScore with the mean of 
# the respective columns.

df['MathScore'] = df['MathScore'].fillna(value=df['MathScore'].mean())
df['EnglishScore'] = df['EnglishScore'].fillna(value=df['EnglishScore'].mean())

df
```

```python
# Encode the Gender column using label encoding (0 for female, 1 for male).

df['Gender'] = df['Gender'].replace({'Female': 0, 'Male': 1})

df
```

```python
# Correct the TotalScore column.

df['TotalScore'] = df['MathScore'] + df['EnglishScore']

df
```

```python
# Standardize the MathScore, EnglishScore, and TotalScore columns using z-score
# normalization.

# z-Score method
# scaled_valued = (value - mean) / std deviation


# copy the DataFrame

scaled_data = df.copy()

# Applying z-score normalization 

scaled_data['MathScore'] = (scaled_data['MathScore'] - scaled_data['MathScore'].mean()) / scaled_data['MathScore'].std()
scaled_data['EnglishScore'] = (scaled_data['EnglishScore'] - scaled_data['EnglishScore'].mean()) / scaled_data['EnglishScore'].std()
scaled_data['TotalScore'] = (scaled_data['TotalScore'] - scaled_data['TotalScore'].mean()) / scaled_data['TotalScore'].std()

scaled_data = scaled_data.round(2)

scaled_data
```
