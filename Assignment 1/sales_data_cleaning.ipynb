```python
import pandas as pd

df = pd.read_csv('./Sales.csv')

df
```

```python
# Creating a new column Revenue by multiplying Price and QuantitySold.

df2 = df.copy()

df2['Revenue_New'] = df2['Price'] * df2['QuantitySold']

df2
```

```python
# Removing outliers in the Price column using the z-score method.

df3 = df2.copy()

# Calculating z-score of 'Price'

df3['Price_z-score'] = (df3['Price'] - df3['Price'].mean()) / df3['Price'].std()

df3 = df3[(df3['Price_z-score'].abs() <= 3)]

df3
```

```python

# Normalize the Price and Revenue columns using Min-Max scaling.

df3['PriceScaled'] = (df3['Price'] - df3['Price'].mean()) / (df3['Price'].max() - df3['Price'].min())
df3['RevenueScaled'] = (df3['Revenue'] - df3['Revenue'].mean()) / (df3['Revenue'].max() - df3['Revenue'].min())


df3
```
